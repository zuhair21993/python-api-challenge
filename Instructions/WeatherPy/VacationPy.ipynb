```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
file = pd.read_csv("../output_data/cities.csv")
vacation_df = pd.DataFrame(file)
vacation_df = vacation_df.rename(columns={"Unnamed: 0": "City ID"})
vacation_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = vacation_df[["Lat", "Lng"]]
humidity = vacation_df["Humidity"]

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity,
                                )
fig.add_layer(heat_layer)
fig
#fig.savefig("../Images/HeatMap.png")
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
summarized_df = vacation_df
summarized_df = summarized_df.loc[(summarized_df["Max Temp"] >= 70) & (summarized_df["Max Temp"] <= 80)]
summarized_df = summarized_df.loc[summarized_df["Wind Speed"] <= 10]
summarized_df = summarized_df.loc[summarized_df["Cloudiness"] == 0]
summarized_df = summarized_df.drop(["City ID"], axis=1)
summarized_df.reset_index(drop=True, inplace=True)
summarized_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = summarized_df
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
for index, row in hotel_df.iterrows():
    params = {
    "types": "Hotel",
    "keyword": "Lodging",
    "radius": 5000,
    "key": g_key,
    }
    
    lat = row["Lat"]
    lng = row["Lng"]
    location = f"{lat}, {lng}"
    params["location"] = location
    #print(params)
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    query_url = requests.get(url, params)
    response = query_url.json()
    hotel_df.loc[index, "Hotel Name"] = response["results"][0]["name"]
    
#pprint(response)
```

```python
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
#fig = gmaps.figure()
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)
#fig.savefig("../Images/Markers.png")
fig

# Display Map
```
